```python
def operation(img, nn_layer):
    writer = SummaryWriter(f'/Users/arponbiswas/Computer-Vision-Projects/Image_classification_projects/Ant_Bee_Image_Classification/TensorBoard-Graph/conv_testing/{nn_layer.name}')
    trans = transforms.Compose([
        transforms.ToTensor()
    ])
    trans_img = trans(img).unsqueeze(0)
    out_img = nn_layer(trans_img).squeeze(0)
    writer.add_graph(nn_layer, trans_img)
    writer.add_image(nn_layer.name, out_img)
    writer.close()
```
