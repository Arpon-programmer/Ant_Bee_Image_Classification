```python
import torch
import torch.nn as nn
from torchsummary import summary
```

```python
class Model_1(nn.Module):
    def __init__(self):
        super(Model_1, self).__init__()
        self.name = 'Model_1'

        self.conv1 = nn.Conv2d(in_channels=3, out_channels=3, kernel_size=1, stride=1, padding=0)

        self.flatten = nn.Flatten()

        self.fc1 = nn.Linear(3*224*224, 1)
    def forward(self, x):
        x = self.conv1(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = torch.sigmoid(x)
        return x

model_1 = Model_1()
summary(model_1, input_size=(3, 224, 224))
# Extended Model with model fc layers
class Model_1_e(nn.Module):
    def __init__(self):
        super(Model_1_e, self).__init__()
        self.name = 'Model_1_e'

        self.conv1 = nn.Conv2d(in_channels=3, out_channels=3, kernel_size=1, stride=1, padding=0)

        self.flatten = nn.Flatten()

        self.fc1 = nn.Linear(3*224*224, 1000)
        self.fc2 = nn.Linear(1000, 50)
        self.fc3 = nn.Linear(50, 1)
    def forward(self, x):
        x = self.conv1(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        x = torch.sigmoid(x)
        return x

model_1_e = Model_1_e()
summary(model_1_e, input_size=(3, 224, 224))
```
